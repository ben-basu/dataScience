```python
#building our model

#using linear regression
```

```python
#importing the required library

from sklearn.linear_model import LinearRegression
```

```python
lreg=LinearRegression()
```

```python
data
```

```python
data.shape
```

```python
#making the test and the train dataset

#using the first 1000 rows to train the model and rest of the rows for testing purpose
```

```python
train=data[0:1000]
```

```python
test=data[1001:]
```

```python
train
```

```python
test
```

```python
#since there are no categorical variable we don't need to make dummy variables
```

```python
#quality is our target variable

#separating the independent and the dependent variables
```

```python
x_train=train.drop('quality',1)
y_train=train['quality']
x_test=test.drop('quality',1)
true_p=test['quality']
```

```python
#fitting the model

lreg.fit(x_train,y_train)
```

```python
#so our model is fitted and now we need to make predictions

predict=lreg.predict(x_test)
```

```python
predict
```

```python
predict=predict.round()
```

```python
predict
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
rms = sqrt(mean_squared_error(true_p, predict))
```

```python
rms
```
